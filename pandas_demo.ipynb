```python
# Average premium in a region
data.groupby(['region'])['charges'].mean()

# select region, average(charges) on a data grouped by region
```

```python
# Average premium in a gender
data.groupby(['gender'])['charges'].mean()
```

```python
# Average premium in a smoker
data.groupby(['smoker'])['charges'].mean()
```

```python
# Average premium based on region, gender and smoker level
data.groupby(['region', 'gender','smoker'])['charges'].mean()
```

```python
# convert a groupby data set into a data from
grp_data = data.groupby(['region', 'gender','smoker'])['charges'].mean().reset_index()
grp_data
```

```python
# sorting
grp_data.sort_values(by=['charges'])
```

```python
# sorting in ascending order
grp_data.sort_values(by=['charges'], ascending=True)
```

```python
# sorting in descending order
grp_data.sort_values(by=['charges'], ascending=False)
```

```python
# Modify the dataframe, use inplace=True
grp_data.sort_values(by=['charges'], inplace=True)
grp_data
```

```python
# loc -> return the dataframe based on your actual index
grp_data.loc[0]
```

```python
# iloc -> return the dataframe based on your current dataset or returns data at physical position
grp_data.iloc[0]
```
