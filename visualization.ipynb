```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x= np.arange(0,10,0.1)
x
```

```python
# line equation, y = mx + c
y = 2*x + 5
y
```

```python
# line chart
plt.plot(x,y)
plt.show()
```

```python
# Customize line chart
# linewidth => define how thick or thin line you need
# linestyle => you can define the style of different lines in your chart. Eligible values: -,:,-.
# color => defines color of the line.
# alpha => degree of transparency/opacity. value ranges from 0 to 1, 0 => transparent, 1=> fully opaque
plt.plot(x,y, linewidth=2, linestyle='-.', color='cyan', alpha=0.2)
plt.show()
```

```python
# There are different functions available for matplotlib as well like
# title => where you can define the title of your chart
# xlabel => define label for x-axis
# ylabel => define label for y-axis
# legend => to define the meaning of lines in your chart. It accepts 2 params,
# list and loc => location where you want to see this legend on the graph
# figure => figsize determines the size of your chart.
plt.figure(figsize=(8,5)) # always at the top
plt.plot(x,y, linewidth=2, linestyle='-.', color='red', alpha=0.5)
plt.title("Matplotlib demo chart")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(['Line 1'], loc='best')
plt.show()
```

```python
# Multiple line charts
x = np.arange(0,10)
y1 = 2*x + 5
y2 = 3*x + 10
```

```python
#Create two or more subplots in the figure
plt.subplot(1,2,1) # (num of rows, num of columns, subplot index in the chart)
plt.plot(x, y1)
plt.title('Graph 1')

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title('Graph 2')

plt.show()
```
